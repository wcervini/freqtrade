```python
from talib.abstract import * 
import numpy as np
import pandas as pd


def Getdata(symbol: str,period:str):
    df = pd.read_json("user_data/data/binance/"+ symbol +"_USDT-"+ period +".json")
    df.columns = ['date', 'open', 'high', 'low', 'close', 'volume']
    df.date = pd.to_datetime(df.date, unit="ms")
    df.set_index("date", inplace=True)
    return df

def valores(df):
    df1=pd.DataFrame(SMA(df,10))
    df1['sma20'] = SMA(df,20)
    df1['sma30'] = SMA(df,30)
    df1['sma50'] = SMA(df,50)
    df1['sma100'] = SMA(df,100)
    df1['sma200'] = SMA(df,200)
    df1['ema10'] = EMA(df,10)
    df1['ema20'] = EMA(df,20)
    df1['ema30'] = EMA(df,30)
    df1['ema50'] = EMA(df,50)
    df1['ema100'] = EMA(df,100)
    df1['ema200'] = EMA(df,200)
    macd = MACD(df)
    df1 = pd.concat([df1,macd])
    return df1    

df = Getdata("ADA","15m")
df_values= valores(df)
df_values
```
